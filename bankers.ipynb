```python
def bankers_safety():

    n = int(input("Enter number of processes: "))
    m = int(input("Enter number of resources: "))

    # total resources
    total_resources = list(map(int, input("\nEnter total resources in system: ").split()))

    max_matrix = []
    print("\nEnter Max Matrix:")

    for i in range(n):
        row = list(map(int, input(f"P{i}: ").split()))

        for j in range(m):
            if row[j] > total_resources[j]:
                print(f"ERROR: Max demand exceeds total resources for Process P{i}")
                return

        max_matrix.append(row)

    allocation = []
    print("\nEnter Allocation Matrix:")

    for i in range(n):
        row = list(map(int, input(f"P{i}: ").split()))

        for j in range(m):
            if row[j] > max_matrix[i][j]:
                print(f"ERROR: Allocation cannot exceed maximum demand for Process P{i}.")
                return

        allocation.append(row)

    # Compute Available = Total - Allocated
    available = [0] * m

    for j in range(m):
        total_alloc = 0
        for i in range(n):
            total_alloc += allocation[i][j]

        if total_alloc > total_resources[j]:
            print("ERROR: Allocated resources exceed total resources")
            return

        available[j] = total_resources[j] - total_alloc

    #  Need matrix
    need = []
    for i in range(n):
        row = []
        for j in range(m):
            row.append(max_matrix[i][j] - allocation[i][j])
        need.append(row)

    print("\nProcess\tAllocation\tMax\t\tNeed\t\tAvailable")

    work = available.copy()
    finish = [False] * n
    safe_seq = []

    while len(safe_seq) < n:

        found = False

        for i in range(n):

            if not finish[i]:
                can_run = True

                for j in range(m):
                    if need[i][j] > work[j]:
                        can_run = False
                        break

                if can_run:

                    print(f"P{i}\t{allocation[i]}\t{max_matrix[i]}\t{need[i]}\t{work}")

                    for j in range(m):
                        work[j] += allocation[i][j]

                    safe_seq.append(i)
                    finish[i] = True
                    found = True
                    break

        if not found:
            print("\nSystem is in DEADLOCK / UNSAFE state")
            return

    print("\nSystem is in SAFE state")

    print("Safe Sequence:", end=" ")
    for i in range(len(safe_seq)):
        print("P" + str(safe_seq[i]), end="")
        if i != len(safe_seq) - 1:
            print(" -> ", end="")


bankers_safety()
```
