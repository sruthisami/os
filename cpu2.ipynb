```python
# Priority Scheduling (Non-Preemptive)

import random
import matplotlib.pyplot as plt

# Take number of processes from user
n = int(input("Enter number of processes: "))

processes = []

# Generate random arrival, burst, and priority
print("\nRandomly generated Arrival Times (AT), Burst Times (BT), and Priorities:")
for i in range(n):
    at = random.randint(0, 5)     # Arrival time 0–5
    bt = random.randint(1, 6)     # Burst time 1–6
    pr = random.randint(1, 5)     # Priority 1–5 (1 = highest)
    processes.append([f"P{i+1}", at, bt, pr])
    print(f"P{i+1} -> AT: {at}, BT: {bt}, Priority: {pr}")

current_time = 0
completed = 0
timeline = []

total_wt = 0
total_tat = 0

print("\nPID  AT  BT  PR  WT  TAT")

# List to track completed processes
done = [False] * n

while completed < n:
    # Find process with highest priority among arrived ones
    idx = -1
    highest_priority = float('inf')

    for i in range(n):
        if (processes[i][1] <= current_time and not done[i]):
            if processes[i][3] < highest_priority:
                highest_priority = processes[i][3]
                idx = i
            elif processes[i][3] == highest_priority:
                if processes[i][1] < processes[idx][1]:
                    idx = i

    # If no process has arrived, CPU is idle
    if idx == -1:
        current_time += 1
        continue

    pid, at, bt, pr = processes[idx]

    wt = current_time - at
    tat = wt + bt

    total_wt += wt
    total_tat += tat

    print(f"{pid}   {at}   {bt}   {pr}   {wt}   {tat}")

    timeline.append((pid, current_time, bt))
    current_time += bt

    done[idx] = True
    completed += 1

print("\nAverage Waiting Time =", round(total_wt / n, 2))
print("Average Turnaround Time =", round(total_tat / n, 2))

# -------- Gantt Chart --------
fig, ax = plt.subplots()

for task in timeline:
    ax.barh(1, task[2], left=task[1], color='lightgreen', edgecolor='black')
    ax.text(task[1] + task[2]/2, 1, task[0],
            ha='center', va='center', color='black', fontsize=10)

ax.set_xlabel("Time")
ax.set_yticks([])
ax.set_title("Priority Scheduling (Non-Preemptive) - Gantt Chart")
plt.show()
```

```python
# Priority Scheduling - Preemptive

import random
import matplotlib.pyplot as plt

# Number of processes
n = int(input("Enter number of processes: "))

processes = []

# Random Arrival Time (0-5), Burst Time (1-6), Priority (1-5)
# Lower number => higher priority
print("\nRandomly generated Arrival Time, Burst Time & Priority:")
for i in range(n):
    at = random.randint(0, 5)
    bt = random.randint(1, 6)
    pr = random.randint(1, 5)
    processes.append([f"P{i+1}", at, bt, pr])
    print(f"P{i+1} -> AT: {at}, BT: {bt}, Priority: {pr}")

remaining_bt = [p[2] for p in processes]

completed = 0
current_time = 0
prev_process = None
timeline = []

wt_list = [0]*n
tat_list = [0]*n
completion_time = [0]*n

while completed < n:
    idx = -1
    min_pr = float('inf')

    # Select highest priority process among arrived ones
    for i in range(n):
        if processes[i][1] <= current_time and remaining_bt[i] > 0:
            if processes[i][3] < min_pr:
                min_pr = processes[i][3]
                idx = i
            elif processes[i][3] == min_pr:
                if processes[i][1] < processes[idx][1]:  # tie-break by arrival
                    idx = i

    if idx == -1:
        current_time += 1  # CPU idle
        continue

    pid = processes[idx][0]

    # Handle context switch
    if prev_process != pid:
        timeline.append([pid, current_time, 1])
    else:
        timeline[-1][2] += 1

    remaining_bt[idx] -= 1
    current_time += 1
    prev_process = pid

    if remaining_bt[idx] == 0:
        completed += 1
        completion_time[idx] = current_time
        tat_list[idx] = completion_time[idx] - processes[idx][1]
        wt_list[idx] = tat_list[idx] - processes[idx][2]

# Print results
print("\nPID  AT  BT  PR  WT  TAT")
for i in range(n):
    pid, at, bt, pr = processes[i]
    print(f"{pid}   {at}   {bt}   {pr}   {wt_list[i]}   {tat_list[i]}")

print("\nAverage Waiting Time =", round(sum(wt_list)/n, 2))
print("Average Turnaround Time =", round(sum(tat_list)/n, 2))

# --- Gantt Chart ---
fig, ax = plt.subplots()
for task in timeline:
    ax.barh(1, task[2], left=task[1], color='crimson', edgecolor='black')
    ax.text(task[1] + task[2]/2, 1, task[0],
            ha='center', va='center', fontsize=9)

ax.set_xlabel("Time")
ax.set_yticks([])
ax.set_title("Priority Scheduling (Preemptive) - Gantt Chart")
plt.show()
```

```python
# Round Robin Scheduling

import random
import matplotlib.pyplot as plt
from collections import deque

# Number of processes
n = int(input("Enter number of processes: "))
tq = int(input("Enter Time Quantum: "))

processes = []

# Random Arrival Time (0-5) and Burst Time (1-6)
print("\nRandomly generated Arrival & Burst Times:")
for i in range(n):
    at = random.randint(0, 5)
    bt = random.randint(1, 6)
    processes.append([f"P{i+1}", at, bt])
    print(f"P{i+1} -> AT: {at}, BT: {bt}")

# Sort by arrival time
processes.sort(key=lambda x: x[1])

remaining_bt = [p[2] for p in processes]
completion_time = [0]*n
wt_list = [0]*n
tat_list = [0]*n

current_time = 0
queue = deque()
timeline = []
visited = [False]*n
completed = 0

while completed < n:
    # Add arrived processes to queue
    for i in range(n):
        if processes[i][1] <= current_time and not visited[i]:
            queue.append(i)
            visited[i] = True

    if not queue:
        current_time += 1
        continue

    idx = queue.popleft()
    pid, at, bt = processes[idx]

    exec_time = min(tq, remaining_bt[idx])
    timeline.append([pid, current_time, exec_time])

    current_time += exec_time
    remaining_bt[idx] -= exec_time

    # Add newly arrived processes during execution
    for i in range(n):
        if processes[i][1] <= current_time and not visited[i]:
            queue.append(i)
            visited[i] = True

    if remaining_bt[idx] > 0:
        queue.append(idx)
    else:
        completed += 1
        completion_time[idx] = current_time
        tat_list[idx] = completion_time[idx] - at
        wt_list[idx] = tat_list[idx] - bt

# Print results
print("\nPID  AT  BT  WT  TAT")
for i in range(n):
    pid, at, bt = processes[i]
    print(f"{pid}   {at}   {bt}   {wt_list[i]}   {tat_list[i]}")

print("\nAverage Waiting Time =", round(sum(wt_list)/n, 2))
print("Average Turnaround Time =", round(sum(tat_list)/n, 2))

# --- Gantt Chart ---
fig, ax = plt.subplots()
for task in timeline:
    ax.barh(1, task[2], left=task[1], edgecolor='black')
    ax.text(task[1] + task[2]/2, 1, task[0],
            ha='center', va='center', fontsize=9)

ax.set_xlabel("Time")
ax.set_yticks([])
ax.set_title(f"Round Robin Scheduling (TQ = {tq}) - Gantt Chart")
plt.show()
```
