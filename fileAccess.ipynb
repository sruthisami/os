```python
#include <iostream>
#include <vector>
using namespace std;

int main() {
    int n;
    cout << "Enter number of disk locations (>=20): ";
    cin >> n;

    vector<string> disk(n);
    cout << "Enter disk status (0 = free, 1 = used):\n";
    for (int i = 0; i < n; i++) {
        cin >> disk[i];
    }

    cout << "\nInitial Disk Status:\n";
    for (string d : disk) cout << d << " ";
    cout << endl;

    int f;
    cout << "\nEnter number of files (>=3): ";
    cin >> f;

    vector<string> names(f);
    vector<int> sizes(f);

    for (int i = 0; i < f; i++) {
        cout << "Enter file " << i + 1 << " name: ";
        cin >> names[i];
        cout << "Enter blocks required for " << names[i] << ": ";
        cin >> sizes[i];
    }

    cout << "\n===== SEQUENTIAL ALLOCATION =====\n";
    vector<string> temp_disk;

    for (int x = 0; x < f; x++) {
        temp_disk = disk;
        string name = names[x];
        int size = sizes[x];
        vector<int> allocated;

        for (int i = 0; i <= n - size; i++) {
            bool free = true;

            for (int j = i; j < i + size; j++) {
                if (temp_disk[j] != "0") {
                    free = false;
                    break;
                }
            }

            if (free) {
                for (int j = i; j < i + size; j++) {
                    temp_disk[j] = name;
                    allocated.push_back(j);
                }
                break;
            }
        }

        if (!allocated.empty()) {
            cout << name << " -> Blocks: ";
            for (int b : allocated) cout << b << " ";

            cout << "\nDiagram: ";
            for (int i = 0; i < allocated.size(); i++) {
                cout << "[" << allocated[i] << "]";
                if (i != allocated.size() - 1) cout << " -> ";
            }
            cout << endl;
        } else {
            cout << name << " -> Allocation Failed\n";
        }
    }


    cout << "\n===== INDEXED ALLOCATION =====\n";

    for (int x = 0; x < f; x++) {
        temp_disk = disk;
        string name = names[x];
        int size = sizes[x];

        vector<int> free_blocks;

        for (int i = 0; i < n; i++) {
            if (temp_disk[i] == "0") free_blocks.push_back(i);
        }

        if (free_blocks.size() < size + 1) {
            cout << name << " -> Allocation Failed\n";
            continue;
        }

        int inode = free_blocks[0];

        vector<int> allocated;
        for (int i = 1; i <= size; i++) {
            allocated.push_back(free_blocks[i]);
            temp_disk[free_blocks[i]] = name;
        }

        temp_disk[inode] = name + "_INODE";

        cout << name << " -> INODE: " << inode << endl;

        cout << "Blocks: ";
        for (int b : allocated) cout << b << " ";

        cout << "\nDiagram: [" << inode << "] -> ";
        for (int i = 0; i < allocated.size(); i++) {
            cout << allocated[i];
            if (i != allocated.size() - 1) cout << ", ";
        }
        cout << endl;
    }


    cout << "\n===== LINKED ALLOCATION =====\n";

    for (int x = 0; x < f; x++) {
        temp_disk = disk;
        string name = names[x];
        int size = sizes[x];

        vector<int> free_blocks;

        for (int i = 0; i < n; i++) {
            if (temp_disk[i] == "0") free_blocks.push_back(i);
        }

        if (free_blocks.size() < size) {
            cout << name << " -> Allocation Failed\n";
            continue;
        }

        vector<int> allocated;

        for (int i = 0; i < size; i++) {
            allocated.push_back(free_blocks[i]);
            temp_disk[free_blocks[i]] = name;
        }

        cout << name << " -> Blocks: ";
        for (int b : allocated) cout << b << " ";

        cout << "\nDiagram: ";
        for (int i = 0; i < allocated.size(); i++) {
            cout << "[" << allocated[i] << "]";
            if (i != allocated.size() - 1)
                cout << " -> ";
            else
                cout << " -> NULL";
        }
        cout << endl;
    }

    return 0;
}
```
