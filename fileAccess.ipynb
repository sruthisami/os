```python
import random

# ---------------- INPUT ----------------
n = int(input("Enter number of disk locations (>=20): "))

disk = []

print("Enter disk status (0 = free, 1 = used):")

for i in range(n):

    status = input(f"Block {i}: ")

    if status == "0":
        disk.append("0")
    else:
        disk.append("USED")

print("\nInitial Disk Status:")
print(disk)

# ---------------- FILE INPUT ----------------
f = int(input("\nEnter number of files (>=3): "))

names = []
sizes = []

for i in range(f):

    name = input(f"Enter file {i+1} name: ")

    size = int(input(f"Enter blocks required for {name}: "))

    names.append(name)
    sizes.append(size)

# =========================================================
# SEQUENTIAL ALLOCATION
# =========================================================
print("\n===== SEQUENTIAL ALLOCATION =====")

for x in range(f):

    name = names[x]
    size = sizes[x]

    allocated = []

    # search contiguous free blocks
    for i in range(n - size + 1):

        free = True

        for j in range(i, i + size):

            if disk[j] != "0":
                free = False
                break

        if free:

            for j in range(i, i + size):
                allocated.append(j)

            break

    # output
    if allocated:

        print(f"{name} -> Blocks:", allocated)

        print("Diagram:", end=" ")

        for i in range(len(allocated)):

            print(f"[{allocated[i]}]", end="")

            if i != len(allocated) - 1:
                print(" -> ", end="")

        print()

    else:
        print(f"{name} -> Allocation Failed")


# =========================================================
# INDEXED ALLOCATION
# =========================================================
print("\n===== INDEXED ALLOCATION =====")

for x in range(f):

    name = names[x]
    size = sizes[x]

    free_blocks = []

    # collect free blocks
    for i in range(n):

        if disk[i] == "0":
            free_blocks.append(i)

    # need one extra inode block
    if len(free_blocks) < size + 1:

        print(f"{name} -> Allocation Failed")
        continue

    inode = free_blocks[0]

    allocated = []

    # allocate data blocks
    for i in range(1, size + 1):
        allocated.append(free_blocks[i])

    # output
    print(f"{name} -> INODE: {inode}")

    print("Blocks:", allocated)

    print(f"Diagram: [{inode}] -> ", end="")

    for i in range(len(allocated)):

        print(allocated[i], end="")

        if i != len(allocated) - 1:
            print(", ", end="")

    print()


# =========================================================
# LINKED ALLOCATION
# =========================================================
print("\n===== LINKED ALLOCATION =====")

for x in range(f):

    name = names[x]
    size = sizes[x]

    free_blocks = []

    # collect free blocks
    for i in range(n):

        if disk[i] == "0":
            free_blocks.append(i)

    # enough blocks required
    if len(free_blocks) < size:

        print(f"{name} -> Allocation Failed")
        continue

    allocated = []

    # allocate blocks
    for i in range(size):
        allocated.append(free_blocks[i])

    # output
    print(f"{name} -> Blocks:", allocated)

    print("Diagram:", end=" ")

    for i in range(len(allocated)):

        print(f"[{allocated[i]}]", end="")

        if i != len(allocated) - 1:
            print(" -> ", end="")
        else:
            print(" -> NULL", end="")

    print()
```
